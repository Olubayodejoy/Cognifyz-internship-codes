```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df3 = pd.read_csv('/content/ResturantDataset.csv')
```

```python
df3
```

```python
df3.isnull().sum()
```

```python
df3['Cuisines'].fillna(df3['Cuisines'].mode()[0], inplace=True)
```

```python
df3.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
df3['Restaurant Name'] = le.fit_transform(df3['Restaurant Name'])
df3['City'] = le.fit_transform(df3['City'])
df3['Address'] = le.fit_transform(df3['Address'])
df3['Locality'] = le.fit_transform(df3['Locality'])
df3['Locality Verbose'] = le.fit_transform(df3['Locality Verbose'])
df3['Cuisines'] = le.fit_transform(df3['Cuisines'])
df3['Currency'] = le.fit_transform(df3['Currency'])
df3['Has Table booking'] = le.fit_transform(df3['Has Table booking'])
df3['Has Online delivery'] = le.fit_transform(df3['Has Online delivery'])
df3['Is delivering now'] = le.fit_transform(df3['Is delivering now'])
df3['Switch to order menu'] = le.fit_transform(df3['Switch to order menu'])
df3['Rating color'] = le.fit_transform(df3['Rating color'])
df3['Rating text'] = le.fit_transform(df3['Rating text'])
df3
```

```python
x = df3.drop(['Cuisines'], axis=1)
y = df3['Cuisines']
```

```python
x
```

```python
y
```

```python
value_counts = y.value_counts()
print(value_counts)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
```

```python
rf = RandomForestRegressor()
```

```python
rf.fit(x_train, y_train)
```

```python
y_pred_rf = rf.predict(x_test)
```

```python
pd.DataFrame({
    'y test' : y_test,
    'y pred rf' : y_pred_rf
})
```

```python
mse = mean_squared_error(y_test, y_pred_rf)
print(f'Mean Squared Error: {mse}')
```

```python
rmse = np.sqrt(mse)
print(f'The root mean squared error is {rmse}')
```

```python
mae = mean_absolute_error(y_test, y_pred_rf)
print(f'The root mean absolute error is {rmse}')
```

```python
r2 = r2_score(y_test, y_pred_rf)
print(f'R^2 Score: {r2}')
```

```python
y_pred_rf = y_pred_rf.round().astype(int)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy = accuracy_score(y_test, y_pred_rf)
print(f'Accuracy: {accuracy}')
```

```python
from sklearn.metrics import precision_score
```

```python
y_pred = rf.predict(x_test)
```

```python
precision = precision_score(y_test, y_pred_rf, average='weighted')
print(f'precision: {precision}')
```
